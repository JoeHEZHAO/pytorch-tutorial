```python
import numpy as np
import os
from skimage import io

# for dirpath, dnames, fnames in 
data_dir='/Users/zhaohe/workspace/data/lung_training_data'
IMAGE_EXT = ['.png', 'jpeg', 'jpg', 'bmp']

for x in ['train', 'val']:
    for dirpath, dnames, fnames in os.walk(os.path.join(data_dir, x)):
        for f in fnames:
            if f.endswith(tuple(IMAGE_EXT)):
                img = io.imread(os.path.join(dirpath, f))
                w,h = img.shape
#               print(w,h)
                print('number {} image is processing'.format(f))
        
                if w != 192:
                    print('number {} image needs to be pad'.format(f))
                    edge = int((192 - w) / 2)
                    img = np.pad(img, edge, 'edge')
                    io.imsave(os.path.join(os.path.join(dirpath, 'padded'), f), img)
                    print(img.shape)
```
